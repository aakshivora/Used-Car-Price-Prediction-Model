```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 0, n_jobs=-1, max_depth=10, n_estimators=100)
rf.fit(X_train, y_train)
```

```python
print('Mean Absolute Error:', round(metrics.mean_absolute_error(y_test, y_pred),2))
print('Mean Squared Error:', round(metrics.mean_squared_error(y_test, y_pred),2))
print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_absolute_error(y_test, y_pred)),2))
```

```python
from sklearn.tree import export_graphviz
import pydot
from IPython.display import Image
tree = rf.estimators_[0]
export_graphviz(tree, out_file ='tree.dot', feature_names=X_train.columns, rounded=True, precision=1)
(graph, )=pydot.graph_from_dot_file("tree.dot")
Image(graph.create_png())
```

```python
importances = list(rf.feature_importances_)
feature_list = list(X_train.columns)
feature_importances = [sorted((feature, round(importance, 2)) for feature, importance in zip(feature_list, importances))]
plt.style.use('fivethirtyeight')
x_values = list(range(len(importances)))
plt.bar(x_values, importances, orientation = 'vertical')
plt.xticks(x_values, feature_list,rotation='vertical')
plt.ylabel('Importances'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

```python
sorted_importances = [importances[1] for importances in feature_importances]
sorted_features = [importances[0] for importances in feature_importances]
cumulative_importances = np.cumsum(sorted_importances, axis=None, dtype=None, out=None)
plt.plot(x_values, cumulative_importances, 'g-')
plt.hlines(y=0.90, xmin=0, xmax=len(sorted_importances), color='r', linestyles = 'dashed')
plt.xticks(x_values, sorted_features, rotation='vertical')
plt.xlabel('Variable');
plt.ylabel('Cumulative Importance');
plt.title('Cumulative Importances');
```
